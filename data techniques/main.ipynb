```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array

import matplotlib.pyplot as plt

import numpy as np
```

# Image to numpy array and then plot it

```python
# load the image via load_img function
img = load_img('img/sample.jpeg')

# details about the image printed below
print('Variable type: ', type(img))
print('Format: ', img.format)
print('Mode: ', img.mode)
print('Size: ', img.size)

# convert the given image into numpy array
img_numpy_array = img_to_array(img)
print('Image converted to numpy array')
print('Variable type: ', type(img_numpy_array))
print('Data type:', img_numpy_array.dtype)
print('Shape:', img_numpy_array.shape)

plt.imshow(img_numpy_array.astype(np.uint8))
```
